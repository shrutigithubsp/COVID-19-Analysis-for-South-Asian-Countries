```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
indonesia=pd.read_csv('Indonesia.csv')
```

```python
indonesia.head()
```

```python
indonesia.describe()
```

```python
# Assuming 'indonesia' is your DataFrame
indonesia['Observation Date'] = pd.to_datetime(indonesia['Observation Date'], errors='coerce', infer_datetime_format=True)

# Create a new column 'Month' with the month extracted from 'Observation Date'
indonesia['Month'] = indonesia['Observation Date'].dt.month

# Now, 'Observation Date' is kept in the dataframe, and 'Month' is added
print(indonesia)
```

```python
# Store the modified data in a new DataFrame named 'India_new'
indonesia_new = indonesia.copy()

# Display the resulting DataFrame
print(indonesia_new)
```

```python
indonesia=indonesia.drop(columns=['Month'])
```

```python
# Create a mapping of month numbers to seasons
season_mapping = {
    3: 'Summer',
    4: 'Summer',
    5: 'Summer',
    6: 'Summer',
    7: 'Summer',
    8: 'Summer',
    9: 'Summer',
    10: 'Monsoon',
    11: 'Monsoon',
    12: 'Monsoon',
    1: 'Monsoon',
    2: 'Monsoon'
}

# Map the month numbers to seasons and create a new 'Seasons' column
indonesia_new['Seasons'] = indonesia_new['Month'].map(season_mapping)

# Display the DataFrame with the new 'Seasons' column
print(indonesia_new)
```

```python
indonesia_new.head()
```

```python
# Convert 'Observation Date' column to datetime format
indonesia_new['Observation Date'] = pd.to_datetime(indonesia_new['Observation Date'])

# Calculate active cases and add a new column 'Active'
indonesia_new['Active'] = indonesia_new['Confirmed'] - indonesia_new['Death'] - indonesia_new['Recovered']

# Now, 'Active' column is added with the active cases information
print(indonesia_new)
```

```python
seasonal_active = indonesia_new.groupby('Seasons')['Active'].sum()

# Custom bright and dark color palette
colors = ['#ff6666', '#ffd966', '#66b3ff', '#99ff99']

# Create a pie chart with the custom color palette
plt.figure(figsize=(8, 8))
plt.pie(seasonal_active, labels=seasonal_active.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Distribution of Active Cases by Season')
plt.show()
```

```python
# Plot 1: Confirmed cases over time
plt.figure(figsize=(8,4))
sns.lineplot(x='Observation Date', y='Confirmed', data=indonesia_new, marker='o')
plt.title('Confirmed Cases Over Time')
plt.xlabel('Observation Date')
plt.ylabel('Confirmed Cases')
plt.show()
```

```python
# Plot 2: Active cases over time
plt.figure(figsize=(8, 4))
sns.lineplot(x='Observation Date', y='Active', data=indonesia_new, marker='o', color='red')
plt.title('Active Cases Over Time')
plt.xlabel('Observation Date')
plt.ylabel('Active Cases')
plt.show()
```

```python
# Plot 1: Bar chart for Confirmed, Death, Recovered, and Active cases based on seasons
fig, ax = plt.subplots(figsize=(10, 4))
indonesia_new.groupby('Seasons')[['Confirmed', 'Death', 'Recovered', 'Active']].sum().plot(kind='bar', ax=ax)
plt.title('Total Cases Based on Seasons')
plt.xlabel('Seasons')
plt.ylabel('Total Cases')
plt.show()
```
