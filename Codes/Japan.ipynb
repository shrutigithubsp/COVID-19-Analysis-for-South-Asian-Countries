```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
japan=pd.read_csv('Japan.csv')
```

```python
japan.head()
```

```python
japan.describe()
```

```python
# Assuming 'japan' is your DataFrame
japan['Observation Date'] = pd.to_datetime(japan['Observation Date'], errors='coerce', infer_datetime_format=True)

# Create a new column 'Month' with the month extracted from 'Observation Date'
japan['Month'] = japan['Observation Date'].dt.month

# Now, 'Observation Date' is kept in the dataframe, and 'Month' is added
print(japan)
```

```python
# Store the modified data in a new DataFrame named 'India_new'
japan_new = japan.copy()

# Display the resulting DataFrame
print(japan_new)
```

```python
japan=japan.drop(columns=['Month'])
```

```python
# Create a mapping of month numbers to seasons
season_mapping = {
    3: 'Summer',
    4: 'Summer',
    5: 'Summer',
    6: 'Summer',
    7: 'Summer',
    8: 'Monsoon',
    9: 'Monsoon',
    10: 'Monsoon',
    11: 'December',
    12: 'December',
    1: 'December',
    2: 'December'
}

# Map the month numbers to seasons and create a new 'Seasons' column
japan_new['Seasons'] = japan_new['Month'].map(season_mapping)

# Display the DataFrame with the new 'Seasons' column
print(japan_new)
```

```python
japan_new.head()
```

```python
# Convert 'Observation Date' column to datetime format
japan_new['Observation Date'] = pd.to_datetime(japan_new['Observation Date'])

# Calculate active cases and add a new column 'Active'
japan_new['Active'] = japan_new['Confirmed'] - japan_new['Death'] - japan_new['Recovered']

# Now, 'Active' column is added with the active cases information
print(japan_new)
```

```python
seasonal_active = japan_new.groupby('Seasons')['Active'].sum()

# Custom bright and dark color palette
colors = ['#ff6666', '#ffd966', '#66b3ff', '#99ff99']

# Create a pie chart with the custom color palette
plt.figure(figsize=(8, 8))
plt.pie(seasonal_active, labels=seasonal_active.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Distribution of Active Cases by Season')
plt.show()
```

```python
# Plot 1: Confirmed cases over time
plt.figure(figsize=(8,4))
sns.lineplot(x='Observation Date', y='Confirmed', data=japan_new, marker='o')
plt.title('Confirmed Cases Over Time')
plt.xlabel('Observation Date')
plt.ylabel('Confirmed Cases')
plt.show()
```

```python
# Plot 2: Active cases over time
plt.figure(figsize=(8, 4))
sns.lineplot(x='Observation Date', y='Active', data=japan_new, marker='o', color='red')
plt.title('Active Cases Over Time')
plt.xlabel('Observation Date')
plt.ylabel('Active Cases')
plt.show()
```

```python
# Plot 1: Bar chart for Confirmed, Death, Recovered, and Active cases based on seasons
fig, ax = plt.subplots(figsize=(10, 4))
japan_new.groupby('Seasons')[['Confirmed', 'Death', 'Recovered', 'Active']].sum().plot(kind='bar', ax=ax)
plt.title('Total Cases Based on Seasons')
plt.xlabel('Seasons')
plt.ylabel('Total Cases')
plt.show()
```
