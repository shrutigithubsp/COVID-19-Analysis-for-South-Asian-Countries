```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Cases.csv')
```

```python
df
```

```python
df.head()
```

```python
# Descriptive Statistics
print("\nDescriptive Statistics:")
print(df.describe())
```

This code calculates basic descriptive statistics (mean, std, min, max, etc.) for each column in the DataFrame.

```python
# Comparative Analysis
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Total Confirmed Cases', data=df)
plt.title('Total Confirmed Cases by Country')
plt.xticks(rotation=45, ha='right')
plt.show()
```

this code generates a bar plot to compare the total confirmed COVID-19 cases across different countries, making it easy to visually compare the impact of the pandemic in each country.

```python
# Comparative Analysis
total_cases_sorted = df.sort_values(by='Total Confirmed Cases', ascending=False)
total_deaths_sorted = df.sort_values(by='Total Deaths', ascending=False)
total_recovered_sorted = df.sort_values(by='Total Recovered', ascending=False)
```

```python
# Convert 'Total Population' column to numeric after removing commas
df['Total Population'] = df['Total Population'].str.replace(',', '').astype(int)

# Population-adjusted Metrics
df['Cases Per Million'] = df['Total Confirmed Cases'] / (df['Total Population'] / 1e6)
df['Deaths Per Million'] = df['Total Deaths'] / (df['Total Population'] / 1e6)
```

This code calculates two population-adjusted metrics for each country in the DataFrame (df).

```python
# Growth Analysis
plt.figure(figsize=(10,4))
sns.lineplot(x='Country', y='Total Confirmed Cases', data=df, marker='o')
plt.title('Growth of Total Confirmed Cases')
plt.xticks(rotation=45, ha='right')
plt.show()
```

this code generates a line plot to visualize the growth trajectory of total confirmed COVID-19 cases for each country. It allows for a quick assessment of how the number of confirmed cases has evolved over time in different countries. The circular markers help highlight specific data points on the line.

```python
# Fatality Rate Analysis
df['Case Fatality Rate'] = (df['Total Deaths'] / df['Total Confirmed Cases']) * 100
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Case Fatality Rate', data=df)
plt.title('Case Fatality Rate by Country')
plt.xticks(rotation=45, ha='right')
plt.show()
```

this code calculates and visualizes the case fatality rate for each country, providing insights into the proportion of confirmed cases that result in death. The bar plot allows for a comparative analysis of case fatality rates across different countries.

```python
# Time Series Analysis
df_time_series = df.melt(id_vars=['Country'], var_name='Metric', value_name='Value')

# Visualization: Line plot for Time Series Analysis
plt.figure(figsize=(12, 6))
sns.lineplot(x='Country', y='Value', hue='Metric', data=df_time_series, marker='o')
plt.title('Time Series Analysis')
plt.xticks(rotation=45, ha='right')
plt.show()
```

This code performs a time series analysis and visualizes the time series data for different metrics (such as Total Confirmed Cases, Total Deaths, Total Recovered, and Total Population) across countries.

```python
# Adding 'Region' column
df['Region'] = ['East Asia', 'South Asia', 'South Asia', 'South Asia', 'East Asia', 'South Asia', 'East Asia',
                 'East Asia', 'South Asia', 'South Asia']

# Visualization: Bar plot for Regional Comparison
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Total Confirmed Cases', hue='Region', data=df)
plt.title('Regional Comparison of Total Confirmed Cases')
plt.xticks(rotation=45, ha='right')
plt.show()
```

 this code facilitates the comparison of total confirmed COVID-19 cases across countries within different regions. The bar plot with regions distinguished by color provides a visual representation of the regional distribution of total confirmed cases.

```python
# Set Seaborn style
sns.set(style="whitegrid")
# Bar plot for Total Active Cases grouped by Climate with improved style
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Climate', y='Total Confirmed Cases', data=df, palette='viridis')

# Add data labels on each bar
for p in ax.patches:
    ax.annotate(f'{int(p.get_height()):,}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.title('Total Confirmed Cases by Climate', fontsize=16)
plt.xlabel('Climate', fontsize=14)
plt.ylabel('Total Active Cases', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Convert 'Observation Date' column to datetime format
india['Observation Date'] = pd.to_datetime(india['Observation Date'])

# Create a new column 'Month' with the month extracted from 'Observation Date'
india['Month'] = india['Observation Date'].dt.month

# Now, 'Observation Date' is kept in the dataframe, and 'Month' is added
print(india)
```

```python
# Store the modified data in a new DataFrame named 'India_new'
india_new = india.copy()

# Display the resulting DataFrame
print(india_new)
```

```python
india=india.drop(columns=['Month'])
```

```python
print(india)
```

```python
# Create a mapping of month numbers to seasons
season_mapping = {
    3: 'Summer',
    4: 'Summer',
    5: 'Summer',
    6: 'Summer',
    7: 'Summer',
    8: 'Monsoon',
    9: 'Monsoon',
    10: 'Monsoon',
    11: 'Winter',
    12: 'Winter',
    1: 'Winter',
    2: 'Winter'
}

# Map the month numbers to seasons and create a new 'Seasons' column
india_new['Seasons'] = india_new['Month'].map(season_mapping)

# Display the DataFrame with the new 'Seasons' column
print(india_new)
```

```python
india_new.head()
```

```python
india.head()
```

```python
india.tail()
```

```python
# Convert 'Observation Date' column to datetime format
india_new['Observation Date'] = pd.to_datetime(india_new['Observation Date'])

# Calculate active cases and add a new column 'Active'
india_new['Active'] = india_new['Confirmed'] - india_new['Death'] - india_new['Recovered']

# Now, 'Active' column is added with the active cases information
print(india_new)
```

```python
india_new.tail()
```

```python
# Group by Seasons and calculate the sum of Active cases
seasonal_active = india_new.groupby('Seasons')['Active'].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']  # Contrasting colors
plt.pie(seasonal_active, labels=seasonal_active.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Distribution of Active Cases by Season')
plt.show()
```

```python
# Plot 1: Confirmed cases over time
plt.figure(figsize=(8,4))
sns.lineplot(x='Observation Date', y='Confirmed', data=india_new, marker='o')
plt.title('Confirmed Cases Over Time')
plt.xlabel('Observation Date')
plt.ylabel('Confirmed Cases')
plt.show()
```

```python
# Plot 2: Active cases over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='Observation Date', y='Active', data=india_new, marker='o', color='red')
plt.title('Active Cases Over Time')
plt.xlabel('Observation Date')
plt.ylabel('Active Cases')
plt.show()
```

```python
# Plot 1: Bar chart for Confirmed, Death, Recovered, and Active cases based on seasons
fig, ax = plt.subplots(figsize=(10, 4))
india_new.groupby('Seasons')[['Confirmed', 'Death', 'Recovered', 'Active']].sum().plot(kind='bar', ax=ax)
plt.title('Total Cases Based on Seasons')
plt.xlabel('Seasons')
plt.ylabel('Total Cases')
plt.show()
```
